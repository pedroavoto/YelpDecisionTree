```python
import pandas as pd
df = pd.read_csv("Cleveland_ProcessedDataSet.csv")
OH_restaurants_Cat = df
```

```python
import pandas as pd
df = pd.read_json('business.json', lines=True)
df.categories = df.categories.fillna('No category')
df = df.loc[df['categories'].str.contains('Restaurants')]
df = df[df['state'].str.contains('OH')]
df = df[df['city'].str.contains('Cleveland')]

Business_Table_2 = df[['business_id', 'attributes']]
Business_Table_2['attributes'].apply(pd.Series)
Business_Table_2 = pd.concat([Business_Table_2.drop(['attributes'], axis=1), Business_Table_2['attributes'].apply(pd.Series)], axis=1)
```

```python
OH_Business_Table = Business_Table_2.drop(['NoiseLevel', 'GoodForMeal', 'Caters', 'RestaurantsTakeOut', 'Ambience', 'BusinessParking','RestaurantsTableService', 'BikeParking', 'RestaurantsAttire', 'BusinessAcceptsBitcoin','DogsAllowed', 'WheelchairAccessible', 'HappyHour', 'Corkage','BYOBCorkage', 'BYOB', 'GoodForDancing', 'Music', 'CoatCheck','BestNights', 'Smoking', 'DriveThru', 'ByAppointmentOnly', 'AgesAllowed','DietaryRestrictions', 'AcceptsInsurance'], axis=1)
```

```python
OH_merged_attribs_Cats =  OH_Business_Table.merge(OH_restaurants_Cat, on='business_id')
```

```python
OH_merged_attribs_Cats = OH_merged_attribs_Cats.drop(['name', 'address', 'city', 'state', 'postal_code', 'categories'], axis=1)
OH_merged_attribs_Cats
```

```python
#Change all null to False except price
#df2.dropna(subset=['three', 'four', 'five'], how='all')
OH_merged_attribs_Cats = OH_merged_attribs_Cats.dropna()
OH_merged_attribs_Cats
```

```python
#Change WiFi to True and False
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'no'", value = "False")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'no'", value = "False")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'free'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'free'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'paid'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'paid'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "None", value = "False")
```

```python
#Change Alcohol to True and False
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'full_bar'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'beer_and_wine'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'full_bar'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'beer_and_wine'", value = "True")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "u'none'", value = "False")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "'none'", value = "False")
OH_merged_attribs_Cats = OH_merged_attribs_Cats.replace(to_replace = "None", value = "False")
OH_merged_attribs_Cats
```

```python
# Change RestaurantPriceRange2 False Values to 2
OH_merged_attribs_Cats.RestaurantsPriceRange2 = OH_merged_attribs_Cats.RestaurantsPriceRange2.replace(to_replace = "False", value = "2")
```

```python
#Drop Values that does not have range prices
OH_merged_attribs_Cats = OH_merged_attribs_Cats.dropna()
```

```python
#Convert stars column from float to int
OH_merged_attribs_Cats['stars'] = OH_merged_attribs_Cats['stars'] * 10
OH_merged_attribs_Cats['stars']= OH_merged_attribs_Cats['stars'].astype(int)
```

```python
#Backup our DF
OH_LabelEncoder = OH_merged_attribs_Cats
```

```python
# variance of the star column
OH_LabelEncoder.loc[:,"stars"].var()
```

```python
#Highly Rated Column
OH_LabelEncoder.loc[(OH_LabelEncoder.stars <= 40) & (OH_LabelEncoder.review_count < 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars <= 40) & (OH_LabelEncoder.review_count > 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars >= 40) & (OH_LabelEncoder.review_count < 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars >= 40) & (OH_LabelEncoder.review_count > 150), 'HighlyRated'] = 'True' 
OH_LabelEncoder.loc[(OH_LabelEncoder.stars < 40) & (OH_LabelEncoder.review_count <= 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars < 40) & (OH_LabelEncoder.review_count >= 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars > 40) & (OH_LabelEncoder.review_count <= 150), 'HighlyRated'] = 'False'
OH_LabelEncoder.loc[(OH_LabelEncoder.stars > 40) & (OH_LabelEncoder.review_count >= 150), 'HighlyRated'] = 'True' 
OH_LabelEncoder.loc[(OH_LabelEncoder.stars == 40) & (OH_LabelEncoder.review_count == 150), 'HighlyRated'] = 'True'
```

```python
#HighlyRated as Target
Target = OH_LabelEncoder['HighlyRated']
```

```python
#Drop the Business_id column
OH_LabelEncoder = OH_LabelEncoder.drop(['business_id'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le_OutdoorSeating = LabelEncoder()
le_BusinessAcceptsCreditCards = LabelEncoder()
le_RestaurantsDelivery = LabelEncoder()
le_RestaurantsReservations = LabelEncoder()
le_HasTV = LabelEncoder()
le_RestaurantsPriceRange2 = LabelEncoder()
le_RestaurantsGoodForGroups = LabelEncoder()
le_WiFi = LabelEncoder()
le_GoodForKids = LabelEncoder()
le_Alcohol = LabelEncoder()
le_stars = LabelEncoder()
le_review_count = LabelEncoder()
le_is_open = LabelEncoder()
le_TypeCuisine = LabelEncoder()
le_HighlyRated = LabelEncoder()
```

```python
#Create new columns on our dataset using he new le values
OH_LabelEncoder['OutdoorSeating_New'] = le_OutdoorSeating.fit_transform(OH_LabelEncoder['OutdoorSeating'])
OH_LabelEncoder['BusinessAcceptsCreditCards_New'] = le_BusinessAcceptsCreditCards.fit_transform(OH_LabelEncoder['BusinessAcceptsCreditCards'])
OH_LabelEncoder['RestaurantsDelivery_New'] = le_RestaurantsDelivery.fit_transform(OH_LabelEncoder['RestaurantsDelivery'])
OH_LabelEncoder['RestaurantsReservations_New'] = le_RestaurantsReservations.fit_transform(OH_LabelEncoder['RestaurantsReservations'])
OH_LabelEncoder['HasTV_New'] = le_HasTV.fit_transform(OH_LabelEncoder['HasTV'])
OH_LabelEncoder['RestaurantsGoodForGroups_New'] = le_RestaurantsGoodForGroups.fit_transform(OH_LabelEncoder['RestaurantsGoodForGroups'])
OH_LabelEncoder['WiFi_New'] = le_WiFi.fit_transform(OH_LabelEncoder['WiFi'])
OH_LabelEncoder['GoodForKids_New'] = le_GoodForKids.fit_transform(OH_LabelEncoder['GoodForKids'])
OH_LabelEncoder['Alcohol_New'] = le_Alcohol.fit_transform(OH_LabelEncoder['Alcohol'])
#OH_LabelEncoder['TypeCuisine_New'] = le_TypeCusine.fit_transform(OH_LabelEncoder['TypeCuisine'])
OH_LabelEncoder['HighlyRated_New'] = le_HighlyRated.fit_transform(OH_LabelEncoder['HighlyRated'])
OH_LabelEncoder.head()
```

```python
#Create dummy variables for TypeCusine
cuisine_dummies = pd.get_dummies(OH_LabelEncoder.TypeCusine, prefix='Cuisine')
OH_LabelEncoder = pd.concat([OH_LabelEncoder, cuisine_dummies], axis=1)
```

```python
#Drop the original columns in favor of the numerical columns
OH_LabelEncoder = OH_LabelEncoder.drop(['OutdoorSeating', 'BusinessAcceptsCreditCards', 'RestaurantsDelivery', 'RestaurantsReservations', 'HasTV', 'RestaurantsGoodForGroups', 'WiFi', 'GoodForKids', 'Alcohol', 'TypeCusine', 'HighlyRated'], axis=1)
OH_LabelEncoder
```

```python
OH_LabelEncoder.columns
```

```python
#Make a copy before dropping HighlyRated_New, stars and review_count
OH_Complete = OH_LabelEncoder
OH_LabelEncoder = OH_LabelEncoder.drop(['HighlyRated_New'], axis=1)
OH_LabelEncoder = OH_LabelEncoder.drop(['stars'], axis=1)
OH_LabelEncoder = OH_LabelEncoder.drop(['review_count'], axis=1)
OH_LabelEncoder = OH_LabelEncoder.drop(['is_open'], axis=1)
```

```python
OH_LabelEncoder.columns
```

```python
OH_LabelEncoder.head()
X = OH_LabelEncoder
y = Target
```

```python
OH_LabelEncoder.columns
```

```python
# Fit a classification tree with max_depth=5 on all data.
from sklearn.tree import DecisionTreeClassifier

treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X, y)
```

```python
#Get the score
treeclf.score(X, y)
```

```python
#Predict and display the entire vector
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)

treeclf.predict(X)
```
